Introduction to Pandas

```
# Creating a simple python list 
a=[1,4,3,7,5,6,3]
a
```

```
# Convert the list into an pandas object call series
import pandas as pd
a = pd.Series(a)
a
```

```
type(a)
```

Simple series operations such as sorting,sum,max and count

```
a.max()
```

```
a.min()
```

```
a.sort_values()
```

```
a.sort_values(ascending=False)
```

```
a.sort_values(ascending=False,inplace=True)
```

```
a.value_counts()
```
